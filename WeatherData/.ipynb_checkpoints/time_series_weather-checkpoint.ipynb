```python
data = pd.read_csv('./weatherHistory.csv')
```

```python
data_fromated_humidity = data.loc[:,['Formatted Date','Humidity']]
data_fromated_windSpeed = data.loc[:,['Formatted Date','Wind Speed (km/h)']]
data_fromated_visibility = data.loc[:,['Formatted Date','Visibility (km)']]
data_fromated_pressure = data.loc[:,['Formatted Date','Pressure (millibars)']]


data_fromated_humidity['D'] = pd.to_datetime(data_fromated_humidity['Formatted Date'])
data_fromated_windSpeed['D'] = pd.to_datetime(data_fromated_windSpeed['Formatted Date'])
data_fromated_visibility['D'] = pd.to_datetime(data_fromated_visibility['Formatted Date'])
data_fromated_pressure['D'] = pd.to_datetime(data_fromated_pressure['Formatted Date'])

data_fromated_humidity['Time'] = [datetime.time(d) for d in data_fromated_humidity['D']]
data_fromated_windSpeed['Time'] = [datetime.time(d) for d in data_fromated_windSpeed['D']] 
data_fromated_visibility['Time'] = [datetime.time(d) for d in data_fromated_visibility['D']] 
data_fromated_pressure['Time'] = [datetime.time(d) for d in data_fromated_pressure['D']] 

data_fromated_humidity['Date'] = [datetime.date(d) for d in data_fromated_humidity['D']] 
data_fromated_windSpeed['Date'] = [datetime.date(d) for d in data_fromated_windSpeed['D']] 
data_fromated_visibility['Date'] = [datetime.date(d) for d in data_fromated_visibility['D']] 
data_fromated_pressure['Date'] = [datetime.date(d) for d in data_fromated_pressure['D']]

data_fromated_humidity = data_fromated_humidity.drop(columns=['D','Formatted Date'])
data_fromated_windSpeed = data_fromated_windSpeed.drop(columns=['D','Formatted Date'])
data_fromated_visibility = data_fromated_visibility.drop(columns=['D','Formatted Date'])
data_fromated_pressure = data_fromated_pressure.drop(columns=['D','Formatted Date'])


data_fromated_humidity = data_fromated_humidity[data_fromated_humidity['Time']==datetime.time(datetime.strptime('00:00:00','%H:%M:%S'))]
data_fromated_windSpeed = data_fromated_windSpeed[data_fromated_windSpeed['Time']==datetime.time(datetime.strptime('00:00:00','%H:%M:%S'))]
data_fromated_visibility = data_fromated_visibility[data_fromated_visibility['Time']==datetime.time(datetime.strptime('00:00:00','%H:%M:%S'))]
data_fromated_pressure = data_fromated_pressure[data_fromated_pressure['Time']==datetime.time(datetime.strptime('00:00:00','%H:%M:%S'))]

data_fromated_humidity= data_fromated_humidity.drop(columns='Time')
data_fromated_windSpeed= data_fromated_windSpeed.drop(columns='Time')
data_fromated_visibility= data_fromated_visibility.drop(columns='Time')
data_fromated_pressure= data_fromated_pressure.drop(columns='Time')


data_fromated_humidity['Date'] = pd.to_datetime(data_fromated_humidity['Date'])
data_fromated_windSpeed['Date'] = pd.to_datetime(data_fromated_windSpeed['Date'])
data_fromated_visibility['Date'] = pd.to_datetime(data_fromated_visibility['Date'])
data_fromated_pressure['Date'] = pd.to_datetime(data_fromated_pressure['Date'])

data_fromated_humidity = data_fromated_humidity.set_index('Date')
data_fromated_windSpeed = data_fromated_windSpeed.set_index('Date')
data_fromated_visibility = data_fromated_visibility.set_index('Date')
data_fromated_pressure = data_fromated_pressure.set_index('Date')

data_fromated_humidity.sort_index(inplace=True)
data_fromated_windSpeed.sort_index(inplace=True)
data_fromated_visibility.sort_index(inplace=True)
data_fromated_pressure.sort_index(inplace=True)
```

```python
data_fromated_humidity.plot()
data_fromated_windSpeed.plot()
data_fromated_visibility.plot()
data_fromated_pressure.plot()
```

```python
train, test = data_fromated_humidity[1:3200],data_fromated_humidity[3200:]
model = AR(train,lags=365)
model_fit = model.fit()
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
mse = mean_squared_error(test, predictions)
print('MSE',mse)
test_copy = test.copy()
test_copy['predictions'] = predictions.values
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(test)
plt.xlabel('Year')
plt.ylabel('Humidity')
plt.plot(test_copy['predictions'], color='red')
```
